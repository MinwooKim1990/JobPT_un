```python
import numpy as np
import torch
import torch.nn as nn
from typing import List, Dict, Tuple, Optional
from collections import defaultdict
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import AutoModelForCausalLM, AutoTokenizer
import pandas as pd

class MonosemanticNeuronAnalyzer:
    def __init__(self, model_name: str = "gpt2-medium", layer_nums: List[int] = None):
        """
        Initialize the analyzer with a model and specific layers to analyze
        
        Args:
            model_name: Name of the huggingface model to analyze
            layer_nums: List of layer numbers to analyze
        """
        self.model = AutoModelForCausalLM.from_pretrained(model_name)
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        
        # If layer_nums not specified, analyze first 4 layers
        if layer_nums is None:
            layer_nums = list(range(4))
        self.layer_nums = layer_nums
        
        self.activation_records = defaultdict(list)
        self.concept_map = defaultdict(dict)
        
    def get_neuron_activations(self, text: str) -> Dict[int, torch.Tensor]:
        """Get neuron activations for given input text"""
        tokens = self.tokenizer(text, return_tensors="pt")
        activations = {}
        
        def hook_fn(module, input, output, layer_num):
            # Get intermediate layer activations
            activations[layer_num] = output[0].detach()  # [0] for gpt2 hidden states
            
        hooks = []
        for layer_num in self.layer_nums:
            layer = self.model.transformer.h[layer_num]
            hook = layer.register_forward_hook(
                lambda mod, inp, out, ln=layer_num: hook_fn(mod, inp, out, ln)
            )
            hooks.append(hook)
            
        with torch.no_grad():
            self.model(**tokens)
            
        for hook in hooks:
            hook.remove()
            
        return activations
    
    def analyze_concept(self, concept_name: str, examples: List[str]):
        """Analyze neuron responses to examples of a specific concept"""
        for example in examples:
            activations = self.get_neuron_activations(example)
            
            for layer_num, layer_activations in activations.items():
                mean_activations = layer_activations.mean(dim=1)
                self.activation_records[layer_num].append({
                    'concept': concept_name,
                    'text': example,
                    'activations': mean_activations
                })
    
    def compute_selectivity(self, threshold: float = 0.9) -> Dict[int, List[int]]:
        """Compute neuron selectivity scores and identify monosemantic neurons"""
        monosemantic_neurons = {}
        self.selectivity_scores = {}  # Store for visualization
        
        for layer_num in self.layer_nums:
            layer_records = self.activation_records[layer_num]
            
            concept_activations = defaultdict(list)
            for record in layer_records:
                concept_activations[record['concept']].append(
                    record['activations'].numpy()
                )
            
            n_neurons = layer_records[0]['activations'].shape[-1]
            selectivity_scores = np.zeros(n_neurons)
            
            for neuron_idx in range(n_neurons):
                max_mean_activation = 0
                total_mean_activation = 0
                
                for concept, activations in concept_activations.items():
                    mean_act = np.mean([act[0, neuron_idx] for act in activations])
                    max_mean_activation = max(max_mean_activation, mean_act)
                    total_mean_activation += mean_act
                
                if total_mean_activation > 0:
                    selectivity_scores[neuron_idx] = max_mean_activation / total_mean_activation
            
            self.selectivity_scores[layer_num] = selectivity_scores
            monosemantic_neurons[layer_num] = np.where(selectivity_scores > threshold)[0].tolist()
            
        return monosemantic_neurons

    def visualize_selectivity(self, save_path: Optional[str] = None):
        """Visualize neuron selectivity scores across layers"""
        if not hasattr(self, 'selectivity_scores'):
            raise ValueError("Must run compute_selectivity() before visualization")
            
        n_layers = len(self.layer_nums)
        fig, axes = plt.subplots(n_layers, 1, figsize=(15, 4*n_layers))
        if n_layers == 1:
            axes = [axes]
            
        for ax, layer_num in zip(axes, self.layer_nums):
            scores = self.selectivity_scores[layer_num]
            sns.histplot(scores, bins=50, ax=ax)
            ax.set_title(f'Layer {layer_num} Neuron Selectivity Distribution')
            ax.set_xlabel('Selectivity Score')
            ax.set_ylabel('Count')
            
        plt.tight_layout()
        if save_path:
            plt.savefig(save_path)
        plt.show()
        
    def visualize_mean_activations(self, layer_num: int, save_path: Optional[str] = None):
        """Visualize mean activations for each concept across neurons in a specified layer."""
        layer_records = self.activation_records[layer_num]
        
        # Prepare data for visualization
        concept_means = defaultdict(list)
        for record in layer_records:
            concept_means[record['concept']].append(record['activations'].numpy())
        
        mean_activations = {concept: np.mean(np.array(acts), axis=0) for concept, acts in concept_means.items()}
        
        # Create a bar plot for mean activations
        plt.figure(figsize=(12, 6))
        for concept, activations in mean_activations.items():
            plt.bar(range(len(activations)), activations, label=concept)
        
        plt.title(f'Mean Activations for Layer {layer_num}')
        plt.xlabel('Neuron Index')
        plt.ylabel('Mean Activation')
        plt.legend()
        
        if save_path:
            plt.savefig(save_path)
        plt.show()

    def visualize_concept_activations(self, layer_num: int, top_n_neurons: int = 10, save_path: Optional[str] = None):
        """Visualize how top neurons respond to different concepts"""
        layer_records = self.activation_records[layer_num]
        
        # Prepare data for heatmap
        concept_neuron_activations = defaultdict(list)
        concepts = set()
        
        for record in layer_records:
            concept = record['concept']
            concepts.add(concept)
            activations = record['activations'].numpy()[0]
            concept_neuron_activations[concept].append(activations)
            
        # Calculate mean activations per concept
        mean_activations = {}
        for concept in concepts:
            mean_activations[concept] = np.mean(concept_neuron_activations[concept], axis=0)
            
        # Create DataFrame for visualization
        df_data = pd.DataFrame(mean_activations).T
        
        # Select top neurons based on activation variance
        neuron_vars = df_data.var()
        top_neurons = neuron_vars.nlargest(top_n_neurons).index
        df_plot = df_data[top_neurons]
        
        # Create heatmap
        plt.figure(figsize=(12, 8))
        sns.heatmap(df_plot, annot=True, cmap='RdBu_r', center=0)
        plt.title(f'Layer {layer_num} Top {top_n_neurons} Neurons: Concept Activations')
        plt.xlabel('Neuron Index')
        plt.ylabel('Concept')
        
        if save_path:
            plt.savefig(save_path)
        plt.show()

# Example usage
def run_analysis():
    # Initialize the analyzer (using GPT-2 Medium)
    analyzer = MonosemanticNeuronAnalyzer(
        model_name="gpt2-medium",
        layer_nums=[0, 1, 2]
    )
    
    # Concepts and example sentences to test
    concept_examples = {
        "Positive Emotion": [
            "I had a really happy day today",
            "I got some great news",
            "It was the best moment ever"
        ],
        "Negative Emotion": [
            "I'm feeling so sad",
            "I'm really disappointed",
            "I'm so angry"
        ],
        "Weather": [
            "It's a clear day today",
            "It's been raining all day",
            "It's getting colder"
        ]
    }
    
    # Analyze neuron responses for each concept
    for concept, examples in concept_examples.items():
        print(f"Analyzing concept: {concept}")
        analyzer.analyze_concept(concept, examples)
    
    # Identify monosemantic neurons
    monosemantic_neurons = analyzer.compute_selectivity(threshold=0.8)
    
    # Visualize results
    analyzer.visualize_selectivity()
    
    # Visualize mean activations for each layer
    for layer_num in analyzer.layer_nums:
        analyzer.visualize_mean_activations(layer_num)
        analyzer.visualize_concept_activations(layer_num)
    
    return monosemantic_neurons

if __name__ == "__main__":
    monosemantic_neurons = run_analysis()
    print("\nIdentified monosemantic neurons by layer:")
    for layer, neurons in monosemantic_neurons.items():
        print(f"Layer {layer}: {len(neurons)} neurons")
```

```python
import torch
from transformers import AutoModel, AutoTokenizer
import matplotlib.pyplot as plt
import numpy as np

# 1. 모델 및 토크나이저 로드
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# 모델을 평가 모드로 전환(grad 불필요)
model.eval()

# 2. 분석할 단어 리스트 (다양한 토큰)
words = [
    "cat", "dog", "lion", "tiger", "apple", "banana", "orange", 
    "car", "train", "music", "mathematics", "politics", "chair", "table"
]

# 3. 특정 뉴런 인덱스 설정
# DistilBERT의 hidden state shape은 [batch_size, seq_length, hidden_size]
# hidden_size는 768 (distilBERT 기준)
# 예: 특정 뉴런 인덱스(예: 200번 뉴런) 선택
neuron_index = 200

# 단순화를 위해 각 단어만 한 문장씩 입력
# [CLS], [SEP] 토큰 포함 -> 실제 단어 토큰은 보통 첫번째 위치(1번 인덱스)
activations = []

for w in words:
    # 토큰화
    inputs = tokenizer(w, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs, output_hidden_states=True)
    # hidden_states: tuple (layers), each layer: [batch_size, seq_len, hidden_size]
    # distilBERT는 6개 레이어 존재. 마지막 레이어 hidden state 사용: outputs.last_hidden_state
    # last_hidden_state: [1, seq_len, 768]
    hidden_state = outputs.last_hidden_state
    
    # 첫 번째 실제 단어 토큰은 일반적으로 [CLS] 다음에 위치하므로 인덱스 1 사용
    # 다만 단일 단어 입력 시 seq_len=2([CLS], 단어), distilBERT 토크나이저는
    # CLS:[101], 단어:[???], SEP 없음 by default
    # distilBERT는 Special token: CLS만 앞에 붙여서 입력하므로 입력 단어는 인덱스 1
    # (토큰화 결과에 따라 인덱스 확인 필요, 여기서는 단일 단어 입력이므로 문제 없음)
    activation_value = hidden_state[0, 1, neuron_index].item()
    activations.append(activation_value)

# 4. 시각화
# 단어별 뉴런 활성화 값을 막대그래프로 표현
plt.figure(figsize=(10, 5))
bars = plt.bar(words, activations, color='skyblue')
plt.xticks(rotation=45)
plt.xlabel("Words")
plt.ylabel(f"Activation of Neuron {neuron_index}")
plt.title(f"Activation values for Neuron {neuron_index} across different words")
plt.grid(True)

# 각 막대 위에 활성화값 표시(선택사항)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., 1.02*height,
             f'{height:.2f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()

# 5. 시각적 결론:
# 이 그래프를 보면 어떤 특정 단어군(예: 동물 관련 단어들: cat, dog, lion, tiger)에서 활성화값이 유독 높다면,
# 해당 뉴런이 "동물"과 연관된 반응을 보이는 단서를 얻을 수 있습니다.
# 이 경우, 이 뉴런이 monosemantic하다고 단정짓기엔 부족하지만,
# 개념적으로 "어떤 개념군에 더 큰 반응을 보인다"는 식으로 해석해볼 수 있습니다.
```
